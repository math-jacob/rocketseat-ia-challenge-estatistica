```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
    ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
    ]
}

df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
```

```python
df_faturamento.info()
```

```python
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref'])
df_faturamento.info()
```

## 1) Traga a Média das Vendas

```python
round(df_faturamento['valor'].mean(), 2)
```

## 2) Criar um gráfico de barras vertical mostrando o mês de referência e o valor

```python
df_faturamento.plot.bar(x='data_ref', y='valor')
```

```python
# Nota-se que o eixo x poluiu a visualização

# criar uma nova coluna "mês" utilizando strftime()
# strftime() É uma função que transforma datas em strings legíveis, usando códigos de formatação
df_faturamento['month_year'] = df_faturamento['data_ref'].dt.strftime('%b/%Y')
df_faturamento.info()
```

```python
df_faturamento.sort_values('data_ref').plot.bar(x='month_year', y='valor')
```

### 3) Criar um gráfico de linhas mostrando o mês de referência e o valor

```python
df_faturamento.plot.line(x='month_year',y='valor')
```
